```python
class Paciente:
    def __init__(self, nombre, especie, edad, motivo):
        self.nombre = nombre
        self.especie = especie
        self.edad = edad
        self.motivo = motivo
        self.siguiente = None
        self.anterior = None


class Veterinaria:
    def __init__(self):
        self.cabeza = None
        self.cola = None

    def agregar_paciente(self, nombre, especie, edad, motivo):
        nuevo_paciente = Paciente(nombre, especie, edad, motivo)
        if not self.cabeza:  # Si la lista est√° vac√≠a
            self.cabeza = self.cola = nuevo_paciente
        else:
            nuevo_paciente.siguiente = self.cabeza
            self.cabeza.anterior = nuevo_paciente
            self.cabeza = nuevo_paciente
        print(f'Paciente "{nombre}" agregado.')

    def mostrar_pacientes_adelante(self):
        actual = self.cabeza
        if not actual:
            print("No hay pacientes registrados.")
            return

        print("Lista de pacientes (inicio ‚Üí fin):")
        while actual:
            print(f'üêæ {actual.nombre}, {actual.especie}, {actual.edad} a√±os, Motivo: {actual.motivo}')
            actual = actual.siguiente
        print()

    def mostrar_pacientes_atras(self):
        actual = self.cola
        if not actual:
            print("No hay pacientes registrados.")
            return

        print("Lista de pacientes (fin ‚Üí inicio):")
        while actual:
            print(f'üêæ {actual.nombre}, {actual.especie}, {actual.edad} a√±os, Motivo: {actual.motivo}')
            actual = actual.anterior
        print()

    def buscar_paciente(self, nombre):
        actual = self.cabeza
        while actual:
            if actual.nombre.lower() == nombre.lower():
                print(
                    f'Paciente encontrado: {actual.nombre}, {actual.especie}, {actual.edad} a√±os, Motivo: {actual.motivo}')
                return actual
            actual = actual.siguiente
        print(f'Paciente "{nombre}" no encontrado.')
        return None

    def eliminar_paciente(self, nombre):
        actual = self.cabeza
        while actual:
            if actual.nombre.lower() == nombre.lower():
                if actual.anterior:
                    actual.anterior.siguiente = actual.siguiente
                else:
                    self.cabeza = actual.siguiente  # Si es el primer nodo

                if actual.siguiente:
                    actual.siguiente.anterior = actual.anterior
                else:
                    self.cola = actual.anterior  # Si es el √∫ltimo nodo

                print(f'üóëÔ∏è Paciente "{nombre}" eliminado.')
                return
            actual = actual.siguiente
        print(f'Paciente "{nombre}" no encontrado.')


def mostrar_menu():
    veterinaria = Veterinaria()

    while True:
        print("\nüêæ MEN√ö VETERINARIA")
        print("1Ô∏è‚É£ Agregar paciente")
        print("2Ô∏è‚É£ Mostrar pacientes (inicio ‚Üí fin)")
        print("3Ô∏è‚É£ Mostrar pacientes (fin ‚Üí inicio)")
        print("4Ô∏è‚É£ Buscar paciente")
        print("5Ô∏è‚É£ Eliminar paciente")
        print("6Ô∏è‚É£ Salir")
        opcion = input("\nSelecciona una opci√≥n: ")

        if opcion == "1":
            nombre = input("Nombre del paciente: ")
            especie = input("Especie del paciente: ")
            edad = input("Edad del paciente: ")
            motivo = input("Motivo de consulta: ")
            veterinaria.agregar_paciente(nombre, especie, edad, motivo)
        elif opcion == "2":
            veterinaria.mostrar_pacientes_adelante()
        elif opcion == "3":
            veterinaria.mostrar_pacientes_atras()
        elif opcion == "4":
            nombre = input("Ingrese el nombre del paciente a buscar: ")
            veterinaria.buscar_paciente(nombre)
        elif opcion == "5":
            nombre = input("Ingrese el nombre del paciente a eliminar: ")
            veterinaria.eliminar_paciente(nombre)
        elif opcion == "6":
            print("¬°Hasta luego!")
            break
        else:
            print("\nOpci√≥n inv√°lida. Intenta de nuevo.")


# Ejecutar el men√∫
mostrar_menu()
```
